```python
# 初始化
import os
import sys
import cx_Oracle 
import pandas as pd

# 添加 oracle 客户端路径
sys.path.append('/Users/kai/instantclient_19_8')

# 创建 DSN
intern = cx_Oracle.makedsn('119.147.213.115', '1522', service_name='wktzorcl')
# 设置工作目录
os.chdir('/Users/kai/Desktop/qs')
def queryOracle(sql):
    # 创建连接
    conn = cx_Oracle.connect(user='wktz_intern01', password='wktz_intern01', dsn=intern)
    # 创建游标
    curs = conn.cursor()
    # 执行查询
    curs.execute(sql)
    # 返回查询结果
    data = curs.fetchall()
    # 获取列名
    columns = [i[0] for i in curs.description]
    # 将结果转换为DataFrame格式
    df = pd.DataFrame(data, columns=columns)
    # 关闭游标和连接
    curs.close()
    conn.close()
    
    return df
```

```python
# 中国石化 测试
sql_1 = '''
SELECT SUBSTR(S_INFO_WINDCODE, 1, 6) as code, a.* FROM WKWD_SYNC.ashareeodprices a
where s_info_windcode like '600028%' and trade_dt > 20230601
order by trade_dt
'''
display(queryOracle(sql_1))
```

```python
# 换手率 from baostock
import baostock as bs

#### 登陆系统 ####
lg = bs.login()
# 显示登陆返回信息
# print('login respond error_code:'+lg.error_code)
# print('login respond  error_msg:'+lg.error_msg)

#### 获取沪深A股历史K线数据 ####
# 详细指标参数，参见“历史行情指标参数”章节；“分钟线”参数与“日线”参数不同。“分钟线”不包含指数。
# 分钟线指标：date,time,code,open,high,low,close,volume,amount,adjustflag
# 周月线指标：date,code,open,high,low,close,volume,amount,adjustflag,turn,pctChg

# 整理股票代码格式： 600000.SH -> sh.600000


def windcode_to_bscode(windcode):
    if windcode[7] == 'S':
        return 'sz.' + windcode[0:6]
    elif windcode[7] == 'Z':
        return 'sh.' + windcode[0:6]
    else:
        return 'sh.' + windcode[0:6]


for i in windcodes['S_INFO_WINDCODE']:
    i_bs = windcode_to_bscode(i)
    rs = bs.query_history_k_data_plus(i_bs,
                                      "date,code,volume,amount,adjustflag,turn,tradestatus,isST",
                                      start_date='2000-01-01', end_date='2023-06-12',
                                      frequency="d", adjustflag="3")
    # print('query_history_k_data_plus respond error_code:'+rs.error_code)
    # print('query_history_k_data_plus respond  error_msg:'+rs.error_msg)

    #### 打印结果集 ####
    data_list = []
    while (rs.error_code == '0') & rs.next():
        # 获取一条记录，将记录合并在一起
        data_list.append(rs.get_row_data())
    result = pd.DataFrame(data_list, columns=rs.fields)

    #### 结果集输出到csv文件 ####
    result.to_csv("/Users/kai/Desktop/qs/turn/{}.csv".format(i), index=False)
    # print(result)

#### 登出系统 ####
bs.logout()
```

```python
# 计算月k数据

for i in windcodes['S_INFO_WINDCODE']:
    os.chdir('/Users/kai/Desktop/qs/data_csv')
    df_month = pd.read_csv('{}.csv'.format(i))
    os.chdir('/Users/kai/Desktop/qs/month')

    # 交易时间变为时间变量
    df_month['TRADE_DT'] = pd.to_datetime(
        df_month['TRADE_DT'], format='%Y%m%d')
    df_month.set_index('TRADE_DT', inplace=True)
    df_month_temp = pd.DataFrame(
        df_month[['S_DQ_OPEN', 'S_DQ_HIGH', 'S_DQ_LOW', 'S_DQ_CLOSE', 'S_DQ_VOLUME', 'S_DQ_AMOUNT']])

    # 月k数据
    df_month_temp = df_month_temp.resample('M').agg(
        {'S_DQ_OPEN': 'first', 'S_DQ_HIGH': 'max', 'S_DQ_LOW': 'min', 'S_DQ_CLOSE': 'last', 'S_DQ_VOLUME': 'sum', 'S_DQ_AMOUNT': 'sum'})
    df_month_temp.columns = ['open', 'high',
                            'low', 'close', 'volume', 'amount']

    # add a column 'mean' = amount / volume
    df_month_temp['mean'] = 10 * \
        df_month_temp['amount'] / df_month_temp['volume']

    df_month_temp.reset_index(inplace=True)
    df_month_temp.to_csv('{}.csv'.format(i), index=False, encoding='utf-8-sig')
```
